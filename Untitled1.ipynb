```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib
import matplotlib.pyplot as plt
import seaborn as sns



model_save_path = '/content/drive/MyDrive/Colab Notebooks/Buyuk_Veri/Models/'

import os
os.makedirs(model_save_path, exist_ok=True)
```

```python
file_path_url_clean = '/content/drive/MyDrive/Colab Notebooks/Buyuk_Veri/Dataset/url_dataset_clean.csv'
df_url = pd.read_csv(file_path_url_clean)

X_url = df_url.drop('Label', axis=1)
y_url = df_url['Label']

X_train_url, X_test_url, y_train_url, y_test_url = train_test_split(X_url, y_url, test_size=0.2, random_state=42, stratify=y_url)

print("URL veri seti yüklendi ve eğitim/test setlerine ayrıldı.")
print(f"URL Eğitim Seti Boyutu: {X_train_url.shape}")
print(f"URL Test Seti Boyutu: {X_test_url.shape}")
```

```python
model_url = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model_url.fit(X_train_url, y_train_url)

joblib.dump(model_url, model_save_path + 'model_url.pkl')

print("URL Modeli başarıyla eğitildi ve kaydedildi.")
```

```python
loaded_model_url = joblib.load(model_save_path + 'model_url.pkl')

y_pred_url = loaded_model_url.predict(X_test_url)
y_proba_url = loaded_model_url.predict_proba(X_test_url)[:, 1]

print("\nURL Modeli Performans Metrikleri:")
print(f"Accuracy: {accuracy_score(y_test_url, y_pred_url):.4f}")
print(f"Precision: {precision_score(y_test_url, y_pred_url):.4f}")
print(f"Recall: {recall_score(y_test_url, y_pred_url):.4f}")
print(f"F1-Score: {f1_score(y_test_url, y_pred_url):.4f}")
print("Sınıflandırma Raporu:\n", classification_report(y_test_url, y_pred_url))

cm_url = confusion_matrix(y_test_url, y_pred_url)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_url, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=['Meşru (Tahmin)', 'Phishing (Tahmin)'],
            yticklabels=['Meşru (Gerçek)', 'Phishing (Gerçek)'])
plt.title('URL Modeli Karmaşıklık Matrisi')
plt.ylabel('Gerçek Sınıf')
plt.xlabel('Tahmin Edilen Sınıf')
plt.show()

fpr_url, tpr_url, _ = roc_curve(y_test_url, y_proba_url)
roc_auc_url = auc(fpr_url, tpr_url)
plt.figure(figsize=(6, 5))
plt.plot(fpr_url, tpr_url, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc_url:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Yanlış Pozitif Oranı (False Positive Rate)')
plt.ylabel('Doğru Pozitif Oranı (True Positive Rate)')
plt.title('URL Modeli ROC Eğrisi')
plt.legend(loc="lower right")
plt.show()
```

```python
file_path_email_clean = '/content/drive/MyDrive/Colab Notebooks/Buyuk_Veri/Dataset/email_dataset_clean.csv'
df_email = pd.read_csv(file_path_email_clean)

tfidf_vectorizer_email = TfidfVectorizer(max_features=5000)
X_email_tfidf = tfidf_vectorizer_email.fit_transform(df_email['content'])

X_email_combined = pd.DataFrame(X_email_tfidf.toarray())

if 'Word_Count' not in df_email.columns:
    df_email['Word_Count'] = df_email['content'].apply(lambda x: len(str(x).strip().split()) if pd.notna(x) else 0)

X_email_combined['Word_Count'] = df_email['Word_Count']

X_email_combined.columns = X_email_combined.columns.astype(str)

y_email = df_email['label']


X_train_email, X_test_email, y_train_email, y_test_email = train_test_split(X_email_combined, y_email, test_size=0.2, random_state=42, stratify=y_email)

print("E-posta veri seti yüklendi ve eğitim/test setlerine ayrıldı.")
print("E-posta Modeli Eğitim Seti Boyutu:", X_train_email.shape)
print("E-posta Modeli Test Seti Boyutu:", X_test_email.shape)
print("X_train_email sütun isimlerinin tipleri kontrol edildi ve düzeltildi.")
```

```python
model_email = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model_email.fit(X_train_email, y_train_email)

joblib.dump(model_email, model_save_path + 'model_email.pkl')
joblib.dump(tfidf_vectorizer_email, model_save_path + 'tfidf_email_vectorizer.pkl')

print("E-posta Modeli ve TF-IDF vektörleştirici başarıyla eğitildi ve kaydedildi.")
```
